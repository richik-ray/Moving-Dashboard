```python
import pandas as pd
pd.set_option('display.max_columns', None)

data = pd.read_csv("data/countypres_2000-2016.csv")
data = data.drop(["state", "FIPS", "office", "candidate", "version"], axis=1)
data = data.dropna()
data = data[data["year"] == 2016]
#data = data[data["party"] in ["democrat", "republican"]]
data["county state"] = data["county"] + ", " + data["state_po"]
```

```python
dataGroup = data.groupby("county state")


def calcPol(obj):
    print(obj[obj["party"] == "republican"])

dataGroup.apply(calcPol)
```

```python
democrat = data[data["party"] == "democrat"]
republican = data[data["party"] == "republican"]
merge = pd.merge(democrat, republican, on=["county state", "year", "state_po", "county", "totalvotes"])
merge["measure"] = (merge["candidatevotes_y"] - merge["candidatevotes_x"])/merge["totalvotes"]
merge = merge.drop(["year", "state_po", "county", "party_x", "candidatevotes_x", "totalvotes", "party_y", "candidatevotes_y"], axis=1)
merge
```

```python
merge.to_csv("Political Parsed.csv", index=False)
```
